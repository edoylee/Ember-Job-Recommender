```python
def update_dist():
    global dist_df
    global next_best_index
    dist_df['indices'] = remains_df['indices']
    dist_df['yes_sum'] = np.sum(yes_dist, axis=1)
    dist_df['no_sum'] = np.sum(no_dist, axis=1)
    dist_df = dist_df.fillna(0)
    dist_df['total_dist'] = remains_df['distances'] + dist_df['yes_sum'] - dist_df['no_sum']
    index_min = dist_df['indices'][dist_df['total_dist'].idxmin()]
    next_best_index = total_df[total_df['indices'] == index_min].index[0]
```

```python
def remove_dups(train, test):
    train_df = pd.read_csv(train, index_col=0)
    test_df = pd.read_csv(test, index_col=0)
    final_jobs = test_df.copy()
    for test_job in test_df.jobs.values:
        for train_job in train_df.jobs.values:
            if test_job == train_job:
                final_jobs.drop(index=final_jobs[final_jobs.jobs == test_job].index[0], inplace=True)
    return final_jobs
```
